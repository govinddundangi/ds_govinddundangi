```python
import pandas as pd
```

```python
fear_greed_path = '/content/drive/My Drive/ds_govinddundangi/csv_files/fear_greed_index.csv'
fear_greed_df = pd.read_csv(fear_greed_path)
print("Fear & Greed Data:")
display(fear_greed_df.head())
```

```python
trader_data_path = '/content/drive/My Drive/ds_govinddundangi/csv_files/historical_data.csv'
trader_df = pd.read_csv(trader_data_path)
print("\nTrader Data:")
display(trader_df.head())
```

```python
fear_greed_df.info()
```

```python
trader_df.info()
```

```python
fear_greed_df['date'] = pd.to_datetime(fear_greed_df['date'])
print("Fear & Greed DataFrame after cleaning:")
fear_greed_df.info()
```

```python
print("\nMissing values in Fear & Greed data:")
print(fear_greed_df.isnull().sum())
```

```python
trader_df['datetime'] = pd.to_datetime(trader_df['Timestamp'], unit='us')
display(trader_df.head())
```

```python
trader_df['Date'] = trader_df['datetime'].dt.date
trader_df['Date'] = pd.to_datetime(trader_df['Date'])
print("Trader DataFrame with new Date column:")
display(trader_df[['Timestamp', 'datetime', 'Date']].head())
```

```python
print("\nMissing values in Trader data:")
print(trader_df.isnull().sum())
```

```python
fear_greed_df.rename(columns={'date': 'Date', 'classification': 'Classification'}, inplace=True)
merged_df = pd.merge(trader_df, fear_greed_df, on='Date', how='left')
print("Successfully Merged DataFrame:")
display(merged_df[['datetime', 'Date', 'Classification', 'Closed PnL']].head())
print("\nMissing sentiment values after merge:")
print(merged_df['Classification'].isnull().sum())
```

```python
merged_df['Closed PnL'] = pd.to_numeric(merged_df['Closed PnL'], errors='coerce')
profitability_summary = merged_df.groupby('Classification')['Closed PnL'].agg(['sum', 'mean', 'median', 'count'])
print("Profitability Summary (Fear vs. Greed):")
display(profitability_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(x=profitability_summary.index, y='mean', data=profitability_summary)
plt.title('Average Profit/Loss per Trade (Fear vs. Greed)', fontsize=16)
plt.ylabel('Average Closed PnL (USD)')
plt.xlabel('Market Sentiment')
output_path = '/content/drive/My Drive/ds_govinddundangi/outputs/avg_pnl_vs_sentiment.png'
plt.savefig(output_path)
plt.show()
```

```python
trader_df['datetime'] = pd.to_datetime(trader_df['Timestamp'], unit='ms')
trader_df['Date'] = trader_df['datetime'].dt.date
trader_df['Date'] = pd.to_datetime(trader_df['Date'])
print("--- Trader Data After Fix ---")
print("Earliest trade date:", trader_df['Date'].min())
print("Latest trade date:", trader_df['Date'].max())
print("\nFirst 5 rows with corrected dates:")
display(trader_df[['datetime', 'Date']].head())
```

```python
merged_df['Size USD'] = pd.to_numeric(merged_df['Size USD'], errors='coerce')
volume_summary = merged_df.groupby('Classification').agg(
    total_volume_usd=('Size USD', 'sum'),
    average_trade_size=('Size USD', 'mean'),
    number_of_trades=('Trade ID', 'count')
)
print("Volume and Activity Summary (Fear vs. Greed):")
display(volume_summary)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=volume_summary.index, y='total_volume_usd', data=volume_summary)
plt.title('Total Trading Volume (Fear vs. Greed)', fontsize=16)
plt.ylabel('Total Volume (USD)')
plt.xlabel('Market Sentiment')
output_path = '/content/drive/My Drive/ds_govinddundangi/outputs/total_volume_vs_sentiment.png'
plt.savefig(output_path)
plt.show()
```

```python
print("--- Checking the raw data ---")
print("\nValue counts for the 'Side' column:")
print(merged_df['Side'].value_counts())
print("\nValue counts for the 'Direction' column:")
print(merged_df['Direction'].value_counts())
print("\n" + "="*50 + "\n")
print("--- Checking our calculated 'Leverage' column ---")
print("Value counts for the 'Leverage' column:")
print(merged_df['Leverage'].value_counts())
```

```python
def assign_leverage(direction):
    if direction == 'Open Long':
        return 1 
    elif direction == 'Open Short':
        return -1
    else:
        return 0  
merged_df['Leverage'] = merged_df['Direction'].apply(assign_leverage)
print("--- Checking 'Leverage' column AFTER the fix ---")
print(merged_df['Leverage'].value_counts())
```

```python
def assign_leverage(direction):
    if direction == 'Open Long':
        return 1 
    elif direction == 'Open Short':
        return -1 
    else:
        return 0 
merged_df['Leverage'] = merged_df['Direction'].apply(assign_leverage)
print("--- Verification: Corrected Leverage Counts ---")
print(merged_df['Leverage'].value_counts())
print("\n" + "="*50 + "\n")
print("--- Data for the Chart (Leverage Summary) ---")
leverage_summary = merged_df.groupby('Classification')['Leverage'].mean().reset_index()
display(leverage_summary)
print("\n" + "="*50 + "\n")
print("--- Final Leverage Chart ---")
plt.figure(figsize=(8, 5))
sns.barplot(x='Classification', y='Leverage', data=leverage_summary)
plt.title('Average Directional Bias (Leverage) vs. Sentiment', fontsize=16)
plt.ylabel('Average Bias (Long=1, Short=-1)')
plt.xlabel('Market Sentiment')
plt.axhline(0, color='black', linewidth=0.8)
output_path = '/content/drive/My Drive/ds_govinddundangi/outputs/average_leverage_vs_sentiment.png'
plt.savefig(output_path)
plt.show()
```
